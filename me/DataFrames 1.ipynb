## ```.rank()```

````python
nba = pd.read_csv("nba.csv").dropna(how = "all")
nba["Salary"] = nba["Salary"].fillna(0).astype("int")
nba.head(3)
````

````python
nba['Weight Rank'] = nba['Weight'].rank(ascending=False).astype('int')
nba
````

````python
nba.sort_values('Weight', ascending=False)
````
