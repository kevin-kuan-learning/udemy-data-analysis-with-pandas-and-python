# Compound Filters

````python
mask1 = df['Gender'] == 'Female'
mask2 = df['Team'] == "Finance"
df[mask1 & mask2].head(10)
````

````python
mask1 = df['Gender'] == 'Female'
mask2 = df['Team'] == "Finance"
df[mask1 | mask2].head(10)
````

````python
mask1 = df['Gender'] == 'Female'
df[~mask1].head(10)
````

````python
mask1 = df['Gender'] == 'Female'
mask2 = df['Team'] == "Finance"
mask3 = df['Last Login Time'] > pd.to_datetime('2021-11-14 12:00:00')
df[~mask1 & (mask2 | mask3)].head(10)
````

# ```between()```

````python
df['Start Date'] = pd.to_datetime(df['Start Date'])
df['Last Login Time'] = pd.to_datetime(df['Last Login Time'])
df['Senior Management'] = df['Senior Management'].astype('bool')
df['Gender'] = df['Gender'].astype('category')
df
````

````python
mask = df['Salary'].between(60000, 70000) # inclusive
df[mask]
````

````python
mask = df['Start Date'].between('1991-01-01', '1991-12-31')
df[mask]
````

````python
mask = df['Last Login Time'].between('8:32:00', '12:00PM')
df[mask]
````

# ```duplicate()```

````python
df['Start Date'] = pd.to_datetime(df['Start Date'])
df['Last Login Time'] = pd.to_datetime(df['Last Login Time'])
df['Senior Management'] = df['Senior Management'].astype('bool')
df['Gender'] = df['Gender'].astype('category')
df
````

````python
df.sort_values(['First Name', 'Team'], inplace=True)
df
````

````python
df['First Name'].duplicated(keep='first')
````

````python
mask = df['First Name'].duplicated(keep='first') # first Aaron
df[~mask]
````

````python
mask = df['First Name'].duplicated(keep='last') # last Aaron
df[~mask]
````

````python
mask = df['First Name'].duplicated(keep=False) # No Aaron
df[~mask]
````

````python
mask = df.duplicated(subset=['First Name', 'Team'], keep='first')
df[mask]
````

````python
df.drop_duplicates(subset=['First Name', 'Team'], keep='first')
````
